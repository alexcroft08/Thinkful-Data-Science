## Playing around with different Tensorflow models and comparing accuracy to computational complexity!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
from tensorflow import keras
# Import the dataset
from keras.datasets import mnist

# Import various componenets for model building
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.layers import LSTM, Input, TimeDistributed
from keras.models import Model
from keras.optimizers import RMSprop


data = keras.datasets.fashion_mnist
(x_img, y_lab),(x_img,y_lab) = data.load_data()
```

```python
(x_img, x_lab),(y_img,y_lab) = data.load_data()
```

```python
print(len(x_img[0]))
```

```python
x_img, y_img = x_img / 255.0 , y_img / 255.0
```

```python
y_img.shape
```

```python
#1
model = tf.keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                         keras.layers.Dense(128,activation='relu'),
                         keras.layers.Dense(10,activation='softmax')])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(x_img,x_lab,epochs=10)

test_loss, test_acc = model.evaluate(y_img,y_lab)

print(test_acc)
```

```python
#2 

model = tf.keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                         keras.layers.Dense(128,activation='relu'),
                         keras.layers.Dropout(0.1),
                         keras.layers.Dense(10,activation='softmax')])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(x_img,x_lab,epochs=10)

test_loss, test_acc = model.evaluate(y_img,y_lab)

print(test_acc)
```

```python
#3
adam = tf.optimizers.Adam(0.001)


model = tf.keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                         keras.layers.Dense(128,activation='relu'),
                         keras.layers.Dropout(0.1),
                         keras.layers.Dense(10,activation='softmax')])

model.compile(optimizer=adam, loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(x_img,x_lab,epochs=10)

test_loss, test_acc = model.evaluate(y_img,y_lab)

print(test_acc)
```

```python
#4
sgd = tf.optimizers.SGD(0.001)


model = tf.keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                         keras.layers.Dense(16,activation='relu'),
                         keras.layers.Dropout(0.1),
                         keras.layers.Dense(10,activation='softmax')])

model.compile(optimizer=sgd, loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.fit(x_img,x_lab,epochs=10)

test_loss, test_acc = model.evaluate(y_img,y_lab)

print(test_acc)
```
