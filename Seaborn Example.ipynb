```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('nba_2017_players_with_salary_wiki_twitter.csv')
```

```python
df.head()
```

```python
g=sns.boxplot(x='RPM',data=df)
```

```python
g = sns.lmplot(x='RPM', y='WINS_RPM', hue='POSITION',data=df)
plt.show()
```

```python
g = sns.lineplot(x='RPM', y='WINS_RPM', hue='POSITION', data=df)
plt.show()
```

```python
g = sns.scatterplot(x='RPM', y='WINS_RPM', hue='POSITION',data=df)
plt.show()
```

```python
g = sns.relplot(x='RPM', y='WINS_RPM', hue='POSITION',data=df)
plt.show()
```

All these plots tell a similar story. The benefit of the 1st plot, the lmplot, is that it shows a line of best fit and confidence interval to the data. However, unlike the scatter and relplots, you are unable to see a lot of the data points as they are on top of each other and jumbled together.
