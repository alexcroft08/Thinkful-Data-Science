```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
weather_df = pd.read_sql_query('select * from weatherinszeged',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()
```

```python
import statsmodels.api as sm

weather_df['tempdifference'] = weather_df['temperature'] - weather_df['apparenttemperature']

Y = weather_df['tempdifference']
X = weather_df[['humidity','windspeed']]
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()

results.summary()
```

The resulting equation is tempdifference = -2.4381 + 3.0292humidity + 0.1193windspeed. 

With all p values lower than 0.001, we can say that both are statistically significant. However, as the humidity coefficient is almost 30 times larger than the windspeed coefficient, we could argue that the windspeed has a much lower effect than humidity on tempdifference, and could be regarded as insignificant to the overall equation.

```python
weather_df['humidityws'] = weather_df['humidity'] * weather_df['windspeed']

Y = weather_df['tempdifference']
X = weather_df[['humidity','windspeed','humidityws']]
X = sm.add_constant(X)
results = sm.OLS(Y,X).fit()

results.summary()
```

The resulting equation is tempdifference = -.0839 - .1775humidity - 0.0905windspeed + .2971humidityws. 


Looking at the p values of the coefficients and their sizes relative to each other, it appears they are all statistically significant. Both the humidity and the windspeed coefficients changed, most notably the humidity coefficient which decreased a lot.
