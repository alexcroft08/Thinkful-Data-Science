# Introduction
When one examines the American Education System, it is impossible to ignore the one major factor that separates it from other countries around the world - cost. According to the College Board in 2018, tuition and fees at private universities have risen to an average of \$34,800 per year. JP Morgan predicts that four year degrees at private institutions could cost as much as \$480,000 by the year 2035. While there is financial aid available to students that fit certain criteria, the majority of students who wish to attain 4 year degrees end up with a substantial amount of debt upon graduation. As of June 2018, Forbes reported that the US student debt total was \$1.5 trillion and that over 44 million people owed debt. The average student debt is \$38,390, and the median ranged between \$10 - 25,000.



That leaves one remaining avenue for students to grab that 4 year degree without accruing a massive amount of loans - scholarship. And for those of us that aren't skilled enough to receive athletic scholarship, which amounts to less than 1% of college students, the goal is to receive academic scholarship. When Admissions departments are reviewing student applications and deciding which to accept, decline, and which to offer scholarship, there are many variables they are looking at. There are two that are often deemed much more important than the rest, and those are the student accumulative Grade Point Average (GPA) from high school and their SAT score. As a result, a considerable amount of time and effort is put in by prospective students to perform on SAT exams in order to maximize their opportunity for acceptance and scholarship into esteemed institutions.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Read in the data set
Scores_df = pd.read_csv('scores.csv')

#Drop rows with incomplete data
Scores_df= Scores_df.dropna()

#Select criteria for schools 
Scores_df = Scores_df[Scores_df['Student Enrollment'] >= 350]


#Create collumn for cumulative SAT score by adding the reading, writing, and math scores
Scores_df['Cumulative Score'] = Scores_df['Average Score (SAT Math)'] + Scores_df['Average Score (SAT Reading)'] + Scores_df['Average Score (SAT Writing)']

#Establish variables of interest for Manhattan, Brooklyn, and the Bronx
Scores_man_m = Scores_df.loc[Scores_df['City'] == 'Manhattan', 'Average Score (SAT Math)']
Scores_man_r = Scores_df.loc[Scores_df['City'] == 'Manhattan', 'Average Score (SAT Reading)']
Scores_man_w = Scores_df.loc[Scores_df['City'] == 'Manhattan', 'Average Score (SAT Writing)']
Scores_man_c = Scores_df.loc[Scores_df['City'] == 'Manhattan', 'Cumulative Score']
Scores_man_m_mean = sum(Scores_man_m) / Scores_man_m.count()
Scores_man_r_mean = sum(Scores_man_r) / Scores_man_r.count()
Scores_man_w_mean = sum(Scores_man_w) / Scores_man_w.count()
Scores_man_c_mean = (Scores_man_m_mean + Scores_man_w_mean + Scores_man_r_mean) / 3

Scores_bronx_m = Scores_df.loc[Scores_df['City'] == 'Bronx', 'Average Score (SAT Math)']
Scores_bronx_r = Scores_df.loc[Scores_df['City'] == 'Bronx', 'Average Score (SAT Reading)']
Scores_bronx_w = Scores_df.loc[Scores_df['City'] == 'Bronx', 'Average Score (SAT Writing)']
Scores_bronx_c = Scores_df.loc[Scores_df['City'] == 'Bronx', 'Cumulative Score']
Scores_bronx_m_mean = sum(Scores_bronx_m) / Scores_bronx_m.count()
Scores_bronx_r_mean = sum(Scores_bronx_r) / Scores_bronx_r.count()
Scores_bronx_w_mean = sum(Scores_bronx_w) / Scores_bronx_w.count()
Scores_bronx_c_mean = (Scores_bronx_m_mean + Scores_bronx_w_mean + Scores_bronx_r_mean) / 3

Scores_brooklyn_m = Scores_df.loc[Scores_df['City'] == 'Brooklyn', 'Average Score (SAT Math)']
Scores_brooklyn_r = Scores_df.loc[Scores_df['City'] == 'Brooklyn', 'Average Score (SAT Reading)']
Scores_brooklyn_w = Scores_df.loc[Scores_df['City'] == 'Brooklyn', 'Average Score (SAT Writing)']
Scores_brooklyn_c = Scores_df.loc[Scores_df['City'] == 'Brooklyn', 'Cumulative Score']
Scores_brooklyn_m_mean = sum(Scores_brooklyn_m) / Scores_brooklyn_m.count()
Scores_brooklyn_r_mean = sum(Scores_brooklyn_r) / Scores_brooklyn_r.count()
Scores_brooklyn_w_mean = sum(Scores_brooklyn_w) / Scores_brooklyn_w.count()
Scores_brooklyn_c_mean = (Scores_brooklyn_m_mean + Scores_brooklyn_w_mean + Scores_brooklyn_r_mean) / 3

#Print out the average values for each variable of interest
print('The average SAT scores from last year are as follows:')
print('Math in Manhattan was {}'.format(Scores_man_m_mean))
print('Reading in Manhattan was {}'.format(Scores_man_r_mean))
print('Writing in Manhattan was {}'.format(Scores_man_w_mean))
print('Cumulative score in Manhattan was {}'.format(Scores_man_c_mean))
print('Math in the Bronx was {}'.format(Scores_bronx_m_mean))
print('Reading in the Bronx was {}'.format(Scores_bronx_r_mean))
print('Writing in the Bronx was {}'.format(Scores_bronx_w_mean))
print('Cumulative score in the Bronx was {}'.format(Scores_bronx_c_mean))
print('Math in Brooklyn last year was {}'.format(Scores_brooklyn_m_mean))
print('Reading in Brooklyn last year was {}'.format(Scores_brooklyn_r_mean))
print('Writing in Brooklyn was {}'.format(Scores_brooklyn_w_mean))
print('Cumulative score in Brooklyn was {}'.format(Scores_brooklyn_c_mean))

#Plot the distribution of variables of interest to examine further
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.hist(Scores_brooklyn_m, alpha=0.7, label='Brooklyn')
plt.hist(Scores_bronx_m, alpha=0.7, label='Bronx')
plt.hist(Scores_man_m, alpha=0.7, label='Manhattan')
plt.legend(loc='upper right')
plt.title('Math Score Distribution')

plt.subplot(2,2,2)
plt.hist(Scores_brooklyn_r, alpha=0.7, label='Brooklyn')
plt.hist(Scores_bronx_r, alpha=0.7, label='Bronx')
plt.hist(Scores_man_r, alpha=0.7, label='Manhattan')
plt.legend(loc='upper right')
plt.title('Reading Score Distribution')

plt.subplot(2,2,3)
plt.hist(Scores_brooklyn_w, alpha=0.7, label='Brooklyn')
plt.hist(Scores_bronx_w, alpha=0.7, label='Bronx')
plt.hist(Scores_man_w, alpha=0.7, label='Manhattan')
plt.legend(loc='upper right')
plt.title('Writing Score Distribution')

plt.subplot(2,2,4)
plt.hist(Scores_brooklyn_c, alpha=0.7, label='Brooklyn')
plt.hist(Scores_bronx_c, alpha=0.7, label='Bronx')
plt.hist(Scores_man_c, alpha=0.7, label='Manhattan')
plt.legend(loc='upper right')
plt.title('Cumulative Score Distribution')
plt.show()
```

### Experimental Hypothesis
This experiment will determine if providing certain schools with technology in the classroom in the form of iPads will improve SAT scores by 8% or more. The data set chosen contains SAT Scores in math, reading, and writing from the public schools in New York City in the Bronx, Manhattan, and Brooklyn boroughs. 

### The method of testing
We will be giving half the schools iPads and then comparing their SAT results with the other half, which is our control group. The variables of interest will be the cumulative SAT score, and we will also examine the individual math, reading, and writing scores to see if there is a significant difference in which, if any, are improving. If the cumulative SAT score in the sample of schools with iPads increases by a margin of 8% or better, we can conclude that the iPads make a significant difference in student SAT scores and they should be installed in every school in NYC.
The selection criteria for which schools receive an iPad and which do not will be random. To qualify for the experiment, schools will need to be similar in size of attendance to assure that the two testing groups are as similar as possible. As a result we will only be looking at schools with an attendance of 350 or more. 

### Rollout Plan
Once the schools have been selected, iPads will be provided to all classrooms for English, Math, and Science courses. Teachers in the classrooms with iPads will be provided one iPad per desk, and the students will be free to utilize them during any down time before/after class, for quizzes, and as the teachers see fit. 1 month into the experiment, all classrooms with the iPads will be given a survey. Students will be asked to respond with how often they use the iPads in downtime and if they believe it is enhancing their learnings. Responses will be recorded as an average for each school on scale of 1-5 and noted. This could provide useful information at the end of the experiment once results are recorded, to see if there is a correlation between schools that are using the iPads more, and results on the SAT.

### Evaluation plan
At the end of the year, SAT scores will be evaluated. If schools with the iPads, on average, increased their cumulative SAT score from the previous year by 8% or better, conclude that the iPads provide significant learning increase and reccomend iPad installation in all schools in NYC. If scores only improve by 4% or less, conclude that their is value in using the iPads but more testing is needed. If scores remain the same or worse, conclude the null hypothesis is true and their is no value in using iPads to improve SAT score performance.
